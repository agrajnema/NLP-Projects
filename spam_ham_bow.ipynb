##### Spam Ham project using Bag of Words (BOW)

```python
import pandas as pd
```

```python
messages = pd.read_csv('./dataset/SMSSpamCollection.csv', sep='\t', names=['label', 'message'])
```

```python
messages.head()
```

##### Data Cleaning and Preprocessing

```python
import nltk
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []

for i in range(0,len(messages)):
    words = messages['message'][i].lower() #change all words to lowercase
    words = re.sub('^a-z', ' ', words) # remove special characters
    words = words.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))] # lemmatize words and remove stopwords
    words = ' '.join(words)
    corpus.append(words)

corpus

```

##### Create Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2000, ngram_range=(1,2))
X = cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
# Convert output categorical feature from spam/ham to 0/1
Y = pd.get_dummies(messages['label'], dtype=int)
```

```python
Y
```

```python
Y = Y['spam'].values
```

```python
Y
```

```python
# Test train split
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=3)
```

```python
# import naiye bayes Machine learning algorithm
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()

# Train the model
nb.fit(X_train, Y_train)
```

```python
# Predict using Naive bayes
Y_pred = nb.predict(X_test)
```
